```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import os

print(f"TensorFlow Version: {tf.__version__}")
```

```python
# Ubah path ini sesuai dengan lokasi folder dataset kamu.
# Karena folder 'dataset' berada di root proyek yang sama dengan tempat Anda menjalankan jupyter,
# maka cukup 'dataset'.
DATA_DIR = 'dataset'
IMAGE_SIZE = (150, 150) # Ukuran gambar akan diubah menjadi 150x150 piksel
BATCH_SIZE = 32 # Jumlah gambar yang diproses dalam satu batch
NUM_CLASSES = 2 # Duku dan Langsat (jumlah kelas)
```

```python
# Augmentasi Data (Untuk meningkatkan variasi dataset)
train_datagen = ImageDataGenerator(
    rescale=1./255,          # Normalisasi piksel dari 0-255 menjadi 0-1
    rotation_range=20,       # Rotasi gambar hingga 20 derajat
    width_shift_range=0.2,   # Geser gambar horizontal
    height_shift_range=0.2,  # Geser gambar vertikal
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Zoom in/out
    horizontal_flip=True,    # Balik gambar horizontal
    fill_mode='nearest',     # Isi piksel yang hilang setelah transformasi
    validation_split=0.2     # 20% data akan digunakan untuk validasi
)

# Generator untuk data training
train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical', # Gunakan 'categorical' untuk klasifikasi multi-kelas
    subset='training'
)

# Generator untuk data validasi
validation_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

# Melihat mapping kelas (penting untuk mengetahui urutan kelas)
print("Class Indices:", train_generator.class_indices)
# Simpan nama kelas untuk penggunaan nanti
class_names = list(train_generator.class_indices.keys())
print("Class Names:", class_names)
```

```python
model = Sequential([
    # Layer Konvolusi pertama
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
    MaxPooling2D((2, 2)),

    # Layer Konvolusi kedua
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Layer Konvolusi ketiga
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Meratakan output dari layer konvolusi menjadi vektor 1D
    Flatten(),

    # Layer Dense (fully connected)
    Dense(512, activation='relu'),
    Dropout(0.5), # Mencegah overfitting dengan menonaktifkan 50% neuron secara acak

    # Layer Output
    Dense(NUM_CLASSES, activation='softmax') # Output layer, softmax untuk probabilitas kelas
])

# Mengkompilasi Model
model.compile(optimizer=Adam(learning_rate=0.001), # Optimizer Adam dengan learning rate 0.001
              loss='categorical_crossentropy',     # Fungsi kerugian untuk klasifikasi multi-kelas
              metrics=['accuracy'])                # Metrik yang akan dipantau selama pelatihan

# Menampilkan ringkasan arsitektur model
model.summary()
```

```python
EPOCHS = 20 # Jumlah iterasi pelatihan. Anda bisa menyesuaikannya.
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator
)
```

```python
# Plot akurasi pelatihan dan validasi
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

# Plot loss pelatihan dan validasi
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# Pastikan folder 'model' sudah ada di root proyek kamu
# Path ini relatif terhadap direktori tempat Anda menjalankan Jupyter (yaitu root proyek)
model_save_path = 'model/duku_langsat_model.h5'
labels_save_path = 'model/labels.txt'

# Simpan model dalam format HDF5
model.save(model_save_path)
print(f"Model saved to {model_save_path}")

# Simpan nama kelas ke file teks. Ini penting agar app.py tahu urutan kelasnya.
with open(labels_save_path, 'w') as f:
    for class_name in class_names:
        f.write(class_name + '\n')
print(f"Labels saved to {labels_save_path}")
```
